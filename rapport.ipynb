# On assigne les variables du problème

# Methode d'Euler
Le graphique qui suit représente la solution de l'équation du problème de Kepler résolue avec la méthode d'Euler. On voit facilement que cette solution est instable, ceci est dû au fait que le pas temporel est trop large. On pourrait facilement améliorer les résultats en augmentant le nombre de pas dans la solution. Cependant, faire ceci augmenterait le temps de calcul.

```python
graph(methode_euler(DX, CONDITIONS_INITIALES, DUREE, NOMBRE_DE_PAS))
```

# Methode du predicteur-correcteur
La méthode du prédicteur correcteur est nettement plus efficace que la méthode d'Euler. Tout d'abord, la solution est stable et elle donne un résultat qui se rapproche de la solution obtenue a l'aide de la fonction scipy.integrate.odeint, et ce avec un nombre de pas somme toute assez petit. Cependant, il est tout de même évident que cette solution est différente de la solution obtenue avec la librairie scipy.

```python
graph(methode_predicteur_correcteur(DX,CONDITIONS_INITIALES,DUREE,NOMBRE_DE_PAS))
```

# Methode de Runge-Kutta d'ordre 2
